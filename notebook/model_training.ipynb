```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

```python
# models

#linear models
from sklearn.linear_model import LinearRegression, Ridge, Lasso

#support vector regressor
from sklearn.svm import SVR

#knn
from sklearn.neighbors import KNeighborsRegressor

#decision trees
from sklearn.tree import DecisionTreeRegressor

#hyper parameter tuning
from sklearn.model_selection import RandomizedSearchCV

#ensemble techniques
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from xgboost import XGBRegressor

#evaluation techniques
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
# tain and test split model
from sklearn.model_selection import train_test_split

#scalar
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder

# column transformer
from sklearn.compose import ColumnTransformer
```

```python
# avoiding warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
from pathlib import Path
file = Path().cwd() / "data" / "stud.csv"

df = pd.read_csv(file)
df.head()
```

```python
# independent and dependent features
X = df.drop("math_score", axis=1)
y = df["math_score"]
```

```python
# train and test split
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=43)
```

```python
cat_cols = X.select_dtypes(include="object").columns
num_cols = X.select_dtypes(exclude="object").columns
```

```python
scaler = StandardScaler()
oh_encoder = OneHotEncoder()

preprocessor = ColumnTransformer(
    transformers=[
        ("cat", oh_encoder, cat_cols),
        ("num", scaler, num_cols)
    ]
)

x_train_new = preprocessor.fit_transform(x_train)
x_test_new  = preprocessor.transform(x_test)
```

```python
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Lasso": Lasso(),
    "Ridge": Ridge(),
    "K-Neighbors Regressor": KNeighborsRegressor(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest Regressor": RandomForestRegressor(),
    "XGBRegressor": XGBRegressor(), 
    "AdaBoost Regressor": AdaBoostRegressor()
}

model_list = []
r2_list =[]

for name,model in models.items():
    # fitting the model
    model.fit(x_train_new, y_train)
    
    # get the prediction
    y_train_pred = model.predict(x_train_new)
    y_test_pred = model.predict(x_test_new)
    
    # evaluation
    mse_train, rmse_train, r2_train = evaluate_model(y_train, y_train_pred)
    mse_test, rmse_test, r2_test = evaluate_model(y_test, y_test_pred)
    
    model_list.append(name)
    
    print(name)
    model_list.append(name)
    
    print('Model performance for Training set')
    print("- Root Mean Squared Error: {:.4f}".format(mse_train))
    print("- Mean Absolute Error: {:.4f}".format(rmse_train))
    print("- R2 Score: {:.4f}".format(r2_train))

    print('----------------------------------')
    
    print('Model performance for Test set')
    print("- Root Mean Squared Error: {:.4f}".format(mse_test))
    print("- Mean Absolute Error: {:.4f}".format(rmse_test))
    print("- R2 Score: {:.4f}".format(r2_test))
    r2_list.append(r2_test)
    
    print('='*35)
    print('\n')
```

```python
pd.DataFrame(list(zip(model_list, r2_list)), columns=['Model Name', 'R2_Score']).sort_values(by=["R2_Score"],ascending=False)
```
